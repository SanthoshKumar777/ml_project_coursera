```python
def displayData(x):
    #getting size of data 'x'
    [m,n]=[x.shape[0],x.shape[1]]
    
    #calculating height and width of each digit
    example_width=int(round(np.sqrt(n)))
    example_height=int(np.ceil(n/example_width))
    
    #caluclting how many rows and columns to display
    disp_rows=np.floor(np.sqrt(m))
    disp_col=np.ceil(m/disp_rows)
    
    disp_array=np.zeros((int(disp_rows*example_height),int(disp_col*example_width)))
    
    curr_ex=0
    for j in range(int(disp_rows)):
        for i in range(int(disp_col)):
            if(curr_ex>m):
                break
            max_val=np.max(np.abs(x[curr_ex,:]))
            disp_array[j*example_height:(j+1)*example_height,i*example_width:((i+1)*example_width)]=np.reshape(x[curr_ex,:],(20,20))/max_val
            curr_ex=curr_ex+1
        if(curr_ex>m):
            break
    plt.figure(figsize=(7,7))
    plt.imshow(disp_array.T)
```

```python
def sigmoid(Z):
    return (1.0/(1.0+np.exp(-Z)))
```

```python
def nnCostFunction(nn_params,input_layer_size,hidden_layer_size,num_labels,x,y,lambda_par):
    #unrolling Parameters from nn_params
    theta1=np.reshape(nn_params[:hidden_layer_size*(input_layer_size+1)],(hidden_layer_size,input_layer_size+1))
    theta2=np.reshape(nn_params[hidden_layer_size*(input_layer_size+1):],(num_labels,hidden_layer_size+1))
    
    #initilization
    m=x.shape[0]
    J=0
    theta1_grad=np.zeros((theta1.shape))
    theta2_grad=np.zeros((theta2.shape))
    delta1=np.zeros((theta1.shape))
    delta2=np.zeros((theta2.shape))
    #feed forward 
    
    [a1,Z2,a2,Z3,a3]=forward_prop(x,theta1,theta2)
    
    out=np.multiply(np.log(a3),y)+np.multiply(np.log(1-a3),(1-y))

    #computting cost
    J=-1*(1/m)*np.sum(out)+(lambda_par/(2*m))*((np.sum(np.power(theta1[:,1:input_layer_size+1],2)))+(np.sum(np.power(theta2[:,1:hidden_layer_size+1],2))))
    
    #backpropagation alogorithm
    
    # From lrCostFunction.m in Ex3
    # err = h - y;
    # grad = (1/m)*(err'*X)' + (lambda/m)*theta;
    # grad(1) = grad(1) - (lambda/m)*theta(1);
    
    for t in range((m)):
        #taking each sample at a time starting with t=0
        #step1
        a1t=a1[t,:]
        Z2t=Z2[t,:]
        a2t=a2[t,:]
        Z3t=Z3[t,:]
        a3t=a3[t,:]
        yt=y[t,:]
    
       
        #Back-propogation algorithm
        #step2    
        del3t= a3t - yt
        Z2t=np.c_[np.ones(1),Z2t]
    
        #step3
        del2t= np.multiply((np.matrix(theta2.T)*del3t.T).T,sigmoidGradient(Z2t))
    
        #step4
        #calucltaing error contribution of each weight
        delta2=delta2+del3t.T*a2t
        delta1=delta1+del2t[:,1:].T*a1t
    #step5
    delta1 = delta1 / m
    delta2 = delta2 / m

    theta1_grad=delta1;
    theta2_grad=delta2;

    # add the gradient regularization term
    delta1[:,1:] = delta1[:,1:] + (theta1[:,1:] * lambda_par) / m
    delta2[:,1:] = delta2[:,1:] + (theta2[:,1:] * lambda_par) / m

    # unravel the gradient matrices into a single array
    grad = np.concatenate((np.ravel(delta1), np.ravel(delta2)))

    return J,grad
```

```python
def forward_prop(x,theta1,theta2):
    a1=np.c_[np.zeros((len(x),1)),x]
    Z2=a1*np.matrix(theta1.T)
    a2=sigmoid(Z2)
    a2=np.c_[np.zeros((a2.shape[0],1)),a2]
    Z3=a2*np.matrix(theta2.T)
    a3=sigmoid(Z3)
    return(a1,Z2,a2,Z3,a3)
```

```python
def sigmoidGradient(Z):
    return(np.multiply(sigmoid(Z),(1-sigmoid(Z))))
```

```python
def ComputeNeumericalGrad(J,theta):
    numgrad = np.zeros(theta.shape)
    perturb = np.zeros(theta.shape)
    e = 0.0001

    for p in range(1, np.size(theta)):
        perturb[p] = e
        loss1 = J(theta - perturb)
        loss2 = J(theta + perturb)

        numgrad[p] = (loss2 - loss1) / (2*e)
        perturb[p] = 0
    return numgrad
```

```python
def RandomIntilaizeWeights(L_in,L_out):
    W=np.zeros((L_in,L_out+1))
    epsilon_init = 0.12
    W=np.random.rand(L_out, 1+L_in) * 2 * epsilon_init -epsilon_init
    return W
```

```python
def Cost(nn_params, input_layer_size, hidden_layer_size, num_labels,x,y, lambda_par):
    #unrolling Parameters from nn_params
    theta1=np.reshape(nn_params[:hidden_layer_size*(input_layer_size+1)],(hidden_layer_size,input_layer_size+1))
    theta2=np.reshape(nn_params[hidden_layer_size*(input_layer_size+1):],(num_labels,hidden_layer_size+1))
    
    #initilization
    m=x.shape[0]
    J=0
    theta1_grad=np.zeros((theta1.shape))
    theta2_grad=np.zeros((theta2.shape))
    delta1=np.zeros((theta1.shape))
    delta2=np.zeros((theta2.shape))
    #feed forward 
    
    [a1,Z2,a2,Z3,a3]=forward_prop(x,theta1,theta2)
    
    out=np.multiply(np.log(a3),y)+np.multiply(np.log(1-a3),(1-y))

    #computting cost
    J=-1*(1/m)*np.sum(out)+(lambda_par/(2*m))*((np.sum(np.power(theta1[:,1:input_layer_size+1],2)))+(np.sum(np.power(theta2[:,1:hidden_layer_size+1],2))))
    return J
```

```python
def Predict(theta1,theta2,x):
    m=x.shape[0]
    num_labels=theta2.shape[0] #no labels 
    p=np.zeros((x.shape[0],1)) #no values to predict=5000
    
    x=np.column_stack((np.zeros((m,1)),x))
    o1=sigmoid(x*np.matrix(theta1).T)                            
    o1=np.column_stack((np.zeros((m,1)),o1))
    h=sigmoid(o1*np.matrix(theta2).T)
    p=np.argmax(h,axis=1)
    p=p+1
    return np.array(p)
```
